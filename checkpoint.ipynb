```python
a="Welcome to the Guess the Number game!\nI'm thinking of a number between 1 and 100. Can you guess what it is?"
print(a)
import random
num = random.randint(1, 100)
while True:
    print("Random Integer:", num)
    guess= input("Enter your guess : ")
    guess=int(guess)
    if  guess==num :
        print("Congratulations! You guessed the number correctly!")
        break
    elif guess<num :
        print("too low!")
    else :
        print("too high!")


```
